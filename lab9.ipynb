```python
#NLP Text mining

from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#sample text dara
text="python is a popular choice for AI devlopment because of its ease of us"
```

```python
#generate a word cloud
wordcloud=WordCloud().generate(text)
```

```python
#display the word cloud using matplotlib
plt.figure(figsize=(10,6))
#plt.imshow(wordcloud,interpolation='hamming') #hamming show color word output
plt.imshow(wordcloud,interpolation='bilinear' )
plt.axis("off")
plt.show()
```
